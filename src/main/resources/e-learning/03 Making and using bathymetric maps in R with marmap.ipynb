## <a name="objective">Objective 

The objective of this module is to get started with marmap a package designed for manipulating
bathymetric data in R.

## <a name="data">Data

This module uses ETOPO1 1 Arc-Minute Global Relief Model dataset developed by the National Geophysical Data Center (NGDC), an office of the National Oceanic and Atmospheric Administration (NOAA). 
ETOPO1 was developed to improve the resolution and accuracy of the ETOPO2v2 2 arc-minute Global Relief
Model, and was designed to support tsunami forecasting, modeling and warning, as well as ocean circulation modeling
and Earth visualization.

## <a name="workflow">Workflow

* Step 1: Load the marmap R package

```r
suppressMessages(library('marmap'))
```

* Step 2: define the area of interest

```r
xmin <- -5
ymin <- 30
xmax <- 35
ymax <- 48
```

* Step 3: Retrieve the bathymetry dataset

```r
medi <- getNOAA.bathy(lon1 = xmin, lon2 = xmax,
lat1 = ymin, lat2 = ymax, resolution = 4)
```

* Step 4: Plot the dataset 

```r
library(repr)
options(repr.plot.width=8, repr.plot.height=6)
options(jupyter.plot_mimetypes = 'image/png')
```

```r
# Creating color palettes
blues <- c("lightsteelblue4", "lightsteelblue3",
"lightsteelblue2", "lightsteelblue1")
greys <- c(grey(0.6), grey(0.93), grey(0.99))
```

```r
plot(medi, image = TRUE, land = TRUE, lwd = 0.03,
bpal = list(c(0, max(medi), greys),
c(min(medi), 0, blues)))
# Add coastline
plot(medi, n = 1, lwd = 0.4, add = TRUE)
```

* Step 5: Plot a transect 

```r
x1 <- 2.2
y1 <- 41.36
x2 <- 8.4
y2 <- 41.36
trsect <- get.transect(medi, x1, y1, x2, y2, distance = TRUE)
```

```r
plotProfile(trsect)
```
